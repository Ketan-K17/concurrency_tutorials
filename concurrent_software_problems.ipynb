## PROBLEMS THAT CAN OCCUR IN CONCURRENT SOFTWARE

### DEADLOCK

Deadlock is a situation where two or more threads / processes are waiting for each other to release a resource, while not being ready to relinquish their own resources

A Good way to picture this: Dining Philosopher's Problem

```python
import threading
import time
import random

class Philosopher(threading.Thread):
    def __init__(self, leftFork, rightFork):
        print("Philosopher Has Sat Down At the Table")
        threading.Thread.__init__(self)
        self.leftFork = leftFork
        self.rightFork = rightFork

    def run(self):
        print("Philosopher: {} has started thinking".format(threading.current_thread()))
        while True:
            time.sleep(random.randint(1,5))
            print("Philosopher {} has finished thinking".format(threading.current_thread()))
            self.leftFork.acquire()
            time.sleep(random.randint(1,5))
            try:
                print("Philosopher {} has acquired the left fork".format(threading.current_thread()))
                self.rightFork.acquire()
                try:
                    print("Philosopher {} has attained both forks, currently eating".format(threading.current_thread()))
                    time.sleep(random.randint(1,5))
                finally:
                    self.rightFork.release()
                    print("Philosopher {} has released the right fork".format(threading.current_thread()))
            finally:
                self.leftFork.release()
                print("Philosopher {} has released the left fork".format(threading.current_thread()))


if __name__ == "__main__":
    # Create 5 forks (locks)
    fork1 = threading.Lock()
    fork2 = threading.Lock()
    fork3 = threading.Lock()
    fork4 = threading.Lock()
    fork5 = threading.Lock()

    forks = [fork1, fork2, fork3, fork4, fork5]
    philosophers = []

    # Create 5 philosophers, each with left and right fork
    # Philosopher 0: forks[0] and forks[1]
    # Philosopher 1: forks[1] and forks[2]
    # etc.
    for i in range(5):
        left_fork = forks[i]
        right_fork = forks[(i + 1) % 5]
        philosopher = Philosopher(left_fork, right_fork)
        philosopher.daemon = True  # Make threads daemon so program can exit
        philosophers.append(philosopher)

    # Start all philosophers
    print("\n=== Starting Dining Philosophers ===\n")
    for philosopher in philosophers:
        philosopher.start()

    # Keep the main thread alive for a bit to observe behavior
    try:
        time.sleep(30)  # Run for 30 seconds
    except KeyboardInterrupt:
        print("\nSimulation interrupted by user")

    print("\nSimulation ended")
```

### RACE CONDITION

A race condition tends to happen when the output of some software depends on the sequence / timing of events that are uncontrollable

Preventing race condition is about synchronizing 'critical sections' of your code (a critical section of your code is one where a shared resource is modified.)

critical sections should be prevented from being executed simultaneously by multiple processes. It should be one at a time.
