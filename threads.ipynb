### LIFE CYCLE OF A THREAD

1. New
2. Runnable
3. Running
4. Not Running / Blocked
5. Dead

```python
import threading
import time
# A very simple method for our thread to execute
def threadWorker():
    # it is only at the point where the thread starts executing
    # that it's state goes from 'Runnable' to a 'Running'
    # state
    print("My Thread has entered the 'Running' State")

    # If we call the time.sleep() method then our thread
    # goes into a not-runnable state. We can do no further work
    # on this particular thread
    time.sleep(10)

    # Thread then completes its tasks and terminates
    print("My Thread is terminating")

# At this point in time, the thread has no state
# it hasn't been allocated any system resources
myThread = threading.Thread(target=threadWorker)

# When we call myThread.start(), Python allocates the necessary system
# resources in order for our thread to run and then calls the thread's
# run method. It goes from 'Starting' state to 'Runnable' but not running
myThread.start()


# Here we join the thread and when this method is called
# our thread goes into a 'Dead' state. It has finished the
# job that it was intended to do.
myThread.join()

print("My Thead has entered a 'Dead' state")


# NOTE: when instantiated, a thread is literally JUST a python object, it's not an actual OS thread yet.
# start() makes a request to the OS to create a new thread, and that OS thread then calls run() internally.
# Calling run() directly bypasses this entirely and just executes on the current thread like a normal method call.
```

### WAYS TO START A THREAD

1. For simple tasks that can be done in a single function, you write the function, and pass it as target to the Thread constructor.

```python
import threading
import time
import random

# task to perform
def executeThread(i):
    print(f"Thread {i} started")
    sleepTime = random.randint(1,10)
    time.sleep(sleepTime)
    print(f"Thread {i} finished executing")

for i in range(10):
    thread = threading.Thread(target=executeThread, args=(i,))
    thread.start()
    print("Active Threads:" , threading.enumerate())
```

2. For complex tasks that can't be grouped in one function, you can subclass the Native Thread Class, and override the run() method to do what you want.

In order for us to define a New Thread that subclasses the native Python thread class, we need to do the following at a bare minimum:

1. Pass in the thread class to our class definition.
2. Call Thread.__init__(self) within our constructor in order for our thread to initialize.
3. Define a run() function that will be called when our thread is started:

```python
from threading import Thread

class myWorkerThread(Thread):
    def __init__(self):
        print("Hello world")
        Thread.__init__(self) # using the native Thread's constructor as is.

    def run(self):
        print("Thread is now running")
        myThread = myWorkerThread()
        print("Created my Thread Object")
        myThread.start()
        print("Started my thread")
        myThread.join()
        print("My Thread finished")
```
