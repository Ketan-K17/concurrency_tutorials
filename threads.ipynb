### LIFE CYCLE OF A THREAD

1. New
2. Runnable
3. Running
4. Not Running / Blocked
5. Dead

```python
import threading
import time
# A very simple method for our thread to execute
def threadWorker():
    # it is only at the point where the thread starts executing
    # that it's state goes from 'Runnable' to a 'Running'
    # state
    print("My Thread has entered the 'Running' State")

    # If we call the time.sleep() method then our thread
    # goes into a not-runnable state. We can do no further work
    # on this particular thread
    time.sleep(10)

    # Thread then completes its tasks and terminates
    print("My Thread is terminating")

# At this point in time, the thread has no state
# it hasn't been allocated any system resources
myThread = threading.Thread(target=threadWorker)

# When we call myThread.start(), Python allocates the necessary system
# resources in order for our thread to run and then calls the thread's
# run method. It goes from 'Starting' state to 'Runnable' but not running
myThread.start()


# Here we join the thread and when this method is called
# our thread goes into a 'Dead' state. It has finished the
# job that it was intended to do.
myThread.join()

print("My Thead has entered a 'Dead' state")


# NOTE: when instantiated, a thread is literally JUST a python object, it's not an actual OS thread yet.
# start() makes a request to the OS to create a new thread, and that OS thread then calls run() internally.
# Calling run() directly bypasses this entirely and just executes on the current thread like a normal method call.
```

### WAYS TO START A THREAD

1. For simple tasks that can be done in a single function, you write the function, and pass it as target to the Thread constructor.

```python
import threading
import time
import random

# task to perform
def executeThread(i):
    print(f"Thread {i} started")
    sleepTime = random.randint(1,10)
    time.sleep(sleepTime)
    print(f"Thread {i} finished executing")

for i in range(10):
    thread = threading.Thread(target=executeThread, args=(i,))
    thread.start()
    print("Active Threads:" , threading.enumerate())
```

2. For complex tasks that can't be grouped in one function, you can subclass the Native Thread Class, and override the run() method to do what you want.

In order for us to define a New Thread that subclasses the native Python thread class, we need to do the following at a bare minimum:

1. Pass in the thread class to our class definition.
2. Call Thread.__init__(self) within our constructor in order for our thread to initialize.
3. Define a run() function that will be called when our thread is started:

```python
from threading import Thread

class myWorkerThread(Thread):
    def __init__(self):
        print("Hello world")
        Thread.__init__(self) # using the native Thread's constructor as is.

    def run(self):
        print("Thread is now running")
        myThread = myWorkerThread()
        print("Created my Thread Object")
        myThread.start()
        print("Started my thread")
        myThread.join()
        print("My Thread finished")
```

### Getting the total number of active threads

There's a function for that

```python
import threading
import time
import random

# task to perform
def executeThread(i):
    print(f"Thread {i} started")
    time.sleep(10)
    print(f"Thread {i} finished executing")

if __name__ == '__main__':
    namelist = ["ketan", "nidhi", "vedant", "nischal", "dakshal", "parag", "siddharth", "rounak", "vaishnavi", "jiya"]
    for i in range(10):
        thread = threading.Thread(target=executeThread, args=(i,), name=namelist[i])
        thread.start()
        print("Active Threads:" , threading.enumerate())
    
    print(f"number of active threads: {threading.active_count()}") # use this to count total active threads. it's the same as enumerate, but enumerate names all of them for you.
```

### Getting the main thread

Every Python program needs at least one thread to run. This one thread that is a part of every Python program is called the main thread.

wherever you are, use threading.main_thread() to call the main thread.

```python
import threading
import time

def myChildThread():
    print("Child Thread Starting")
    time.sleep(5)
    print("Current Thread ----------")
    print(threading.current_thread())
    print("-------------------------")
    print("Main Thread -------------")
    print(threading.main_thread())
    print("-------------------------")
    print("Child Thread Ending")
    
child = threading.Thread(target=myChildThread)
child.start()
child.join()
```

### Identifying threads

Having some sort of order in the naming of threads in your code (based on the task they're trying to achieve) makes debugging much easier. One good approach is to have naming conventions for groups of threads that are doing the same / similar tasks.

```python
import threading
import time

def myThread():
    print("Thread {} starting".format(threading.current_thread().name))
    time.sleep(10)
    print("Thread {} ending".format(threading.current_thread().name))

for i in range(4):
 threadName = "Thread-" + str(i)
 thread = threading.Thread(name=threadName, target=myThread) # naming the threads while creating it.
 thread.start()

print("{}".format(threading.enumerate()))    
```

### Ending Threads

DON'T. Bad practice. A small workaround to this is the fact that you can't end threads, but you can end processes.

```python
from multiprocessing import Process
import time

def myWorker():
    t1 = time.time()
    print("Process started at: {}".format(t1))
    time.sleep(20)

myProcess = Process(target=myWorker)
print("Process {}".format(myProcess))
myProcess.start()
print("Terminating Process...")
myProcess.terminate()
myProcess.join()
print("Process Terminated: {}".format(myProcess))

# note how the program ends immediately instead of lasting 20s.
```
