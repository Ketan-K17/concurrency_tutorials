```python
# FORKING A PROCESS

import os

def child():
    print("We are in the child process with PID= {}".format(os.getpid()))

def parent():
    print("We are in the parent process with PID= {}".format(os.getpid()))
    pid = os.fork()  # fork() returns 0 in the child process, and the child's PID value in the parent process. This now means 2 separate processes are currently running this code, 2 different, independent flows of control will now flow through code, one of child (return value 0 path, and another of parent (return value != 0 path)
    if pid == 0:
        child()
    else:
        print("We are in the parent process and our child process has PID= {}".format(pid))
        
parent()
```
