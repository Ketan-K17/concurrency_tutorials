### 1. Adding full / empty checks

This is to prevent MemoryErrors mostly... You should limit the max size of a queue, always..

```python
import threading
from queue import Queue
import time

def myPublisher(queue: Queue):
    while not queue.full():
        queue.put(1)
        print("{} Appended 1 to queue: {}".format(threading.current_thread(), queue.qsize()))
        time.sleep(1)
    print("Operation failed! Queue's full!")

myQueue = Queue(maxsize=5)
threads = []

for i in range(4):
    thread = threading.Thread(target=myPublisher, args=(myQueue,))
    thread.start()
    threads.append(thread)

for thread in threads:
    thread.join()
```

### 2. Using the join() method

Queue objects have their own join() method (not the same as threading.join() mind you).

Queue.join() blocks until ALL items in the queue have been consumed (gotten and task_doned)
i.e. it unblocks when unfinished_tasks = 0
- whenever new item append (q.put()), unfinished_tasks++
- whenever task_done() called, unfinished_tasks--

note that q.get() doesn't change unfinished_tasks

```python
from threading import Thread, current_thread
from queue import Queue
import time

def mySubscriber(queue: Queue):
    while not queue.empty():
        item = queue.get()
        if item is None:
            break
        print("{} removed {} from the queue".format(current_thread(), item))
        print(f"unfinished tasks before task done: {queue.unfinished_tasks}")
        queue.task_done()
        print(f"unfinished tasks after task done: {queue.unfinished_tasks}")
        time.sleep(1)


myQueue = Queue()
for i in range(5):
    myQueue.put(i)
print("Queue Populated")
print(f"Initial count of unfinished tasks : {myQueue.unfinished_tasks}")
print("Adding one more element...")
myQueue.put(5)
print(f"After q.put(), unfinished tasks : {myQueue.unfinished_tasks}")

thread = Thread(target=mySubscriber, args=(myQueue,))
thread.start()
print("Not Progressing Till Queue is Empty")
myQueue.join() # blocks until all items in queue consumed
print("Queue is now empty")
```

### 3. Using Dequeues

Dequeues are just regular queues, but you can push/pop from front / back as opposed to regular queues, where you can only push_back(), and pop_front()

```python
import collections

dq = collections.deque("123456")
print("Dequeue: {}".format(dq))

for item in dq:
    print("Item {}".format(item))

print("Left Most Element: {}".format(dq[0]))
print("Right Most Element: {}".format(dq[-1]))
```

#### Appending / Popping from Dequeue

you use the append() and appendleft() methods for appending
pop() and popleft() for popping

```python
import collections
dq = collections.deque('123456')

print("Deque: {}".format(dq))
dq.append('1')
print("Deque: {}".format(dq))
dq.pop()
print("Deque: {}".format(dq))
dq.appendleft('6')
print("Deque: {}".format(dq))
dq.popleft()
print("Deque: {}".format(dq))
```

#### Insertion into Dequeue

```python
import collections
dq = collections.deque('123456')
print("Deque: {}".format(dq))
dq.insert(5,'5')
print("Deque: {}".format(dq))
```

#### Dequeue Rotation

```python
import collections
dq = collections.deque('123456')
print("Deque: {}".format(dq))
dq.rotate(3) # shift right
print("Deque: {}".format(dq))
dq.rotate(-2) # shift left
print("Deque {}".format(dq))
```
