```python
# sequential download of images - ~5s time for 10 image downloads.

import urllib.request
import time

def downloadImage(imagePath, fileName):                                  
    print("Downloading Image from ", imagePath)
    urllib.request.urlretrieve(imagePath, fileName)                      

def main():                                                              
    start_time = time.time()

    for i in range(10):
        imageName = "temp/image-" + str(i) + ".jpg"
        import random
        random_id = random.randint(0, 50)
        downloadImage(f"https://picsum.photos/id/{random_id}/200/300", imageName)
                                                                        
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"\nTotal time to download 10 images: {elapsed_time:.2f}seconds")                                                                

if __name__ == '__main__':                                               
    main()  
```

```python
# multi-threaded approach - takes ~0.5s.

import threading
import urllib.request
import time
def downloadImage(imagePath, fileName):
    print("Downloading Image from ", imagePath)
    urllib.request.urlretrieve(imagePath, fileName)
    print("Completed Download")

def executeThread(i):
    imageName = "temp/image-" + str(i) + ".jpg"
    import random
    random_id = random.randint(0, 50)
    downloadImage(f"https://picsum.photos/id/{random_id}/200/300", imageName)

def main():
    t0 = time.time()
    # create an array which will store a reference to
    # all of our threads
    threads = []
    # create 10 threads, append them to our array of threads
    # and start them off
    for i in range(100):
        thread = threading.Thread(target=executeThread, args=(i,))
        threads.append(thread)
        thread.start()

    # ensure that all the threads in our array have completed
    # their execution before we log the total time to complete
    # NOTE: for all threads except daemons, you need to manually join them to the main thread.
    for thread in threads:
        thread.join()


    # calculate the total execution time
    t1 = time.time()
    totalTime = t1 - t0
    print("Total Execution Time {}".format(totalTime))
        
if __name__ == '__main__':
    main()
```

```python
# using asyncio, co-routines instead of threads. A single smart thread which knows how to switch between tasks that require waiting.. Asyncio is perfect for such 'waiting' tasks (IO bound ones especially)

# 0.5s, nearly as fast..

# BUT, for 100 images, multi-threading takes 1.8s, asyncio 0.9s.

import asyncio                                                           
import aiohttp                                                           
import time                                                              
import random                                                            

async def downloadImage(session, imagePath, fileName):
    print("Downloading Image from ", imagePath)
    async with session.get(imagePath) as response:
        with open(fileName, 'wb') as f:                                  
            f.write(await response.read())
    print("Completed Download")                                          
            
async def executeTask(session, i):
    imageName = f"temp/image-{i}.jpg"
    random_id = random.randint(0, 50)                                    
    await downloadImage(session, f"https://picsum.photos/id/{random_id}/200/300", imageName)              
            
async def main():
    t0 = time.time()
                                                                        
    # Use aiohttp for async HTTP requests
    async with aiohttp.ClientSession() as session:                       
        # Create 10 coroutine tasks
        tasks = [executeTask(session, i) for i in range(100)]             

        # Wait for all to complete                                       
        await asyncio.gather(*tasks)

    t1 = time.time()
    totalTime = t1 - t0
    print(f"Total Execution Time: {totalTime:.2f} seconds")              

await main()    


# threads are the old way of doing things.. it's true parallelism, but it doesn't scale.. if i had to download 100 images, threads would really slow down... asyncio can do it pretty easily and quick..

# you use multi-threading for the foll cases only - 
# 1. CPU-bound work: asyncio doesn't help here..
# 2. for non-async libraries..

```
