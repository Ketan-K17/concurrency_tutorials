### Logging

logging wasn't covered as extensively as i hoped in this book... but yes.. I did get the idea that by changing the format of the basicConfig (or any equivalent statement) we can get a lot more info out of our programs. But yeah... needs more work..

```python
import logging
from multiprocessing import Pool

def init_worker():
    logging.basicConfig(filename='myapp.log', level=logging.INFO, format='%(processName)-10s %(asctime)s:%(levelname)s:%(message)s')

def myTask(n):
    logging.info("{} being processed".format(n))
    logging.info("Final Result: {}".format(n*2))
    return n*2

def main():
    with Pool(4, initializer=init_worker) as p:
        p.map(myTask, [2,3,4,5,6,])

if __name__ == '__main__':
    main()
```
