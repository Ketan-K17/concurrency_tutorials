```python
# understanding IO bottlenecks in our code with the help of a sample web-crawler, which first requests a web page, and then traverses / processes it.

import urllib.request
import time
from bs4 import BeautifulSoup

t0 = time.time()
req = urllib.request.urlopen('https://ketankunkalikar.blogspot.com/')
t1 = time.time()

print("Total Time To Fetch Page: {} Seconds".format(t1-t0)) # ~1.4s

soup = BeautifulSoup(req.read(), "html.parser")
for link in soup.find_all('a'):
    print(link.get('href'))

t2 = time.time()
print("Total Execeution Time: {} Seconds".format(t2-t1)) # ~0.1s
```
